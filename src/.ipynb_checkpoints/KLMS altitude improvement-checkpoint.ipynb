```python
%matplotlib inline
from kaf_test import*
import numpy as np
import scipy.io as scio
import sys
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
from __future__ import division
import time
import os
cwd = os.getcwd()
```

# Load data #

```python
delay = 30 # filter delay order
## Electric bicycle data ##
database = 'bicycle_data.mat' # database name
mat = scio.loadmat(database) # data load

x_noise = mat['x_noise'][0] # current signal
y_noise = mat['y_noise'][0] # voltage signal
z_noise = mat['z_noise'][0] # temperature signal
a_noise = mat['a_noise'][0] # altitude signal

simLength = len(x_noise) # data length    
hor = 1# prediction horizon
t_new = np.arange(hor+delay,simLength,1) # time array

noiseSignal = [x_noise,y_noise,a_noise]
noiseSignal = np.array(noiseSignal) # list to ndarray

# Autoregressive signal #
[noiseSignal,regSignal] = delaySignal(noiseSignal,delay)

regSignal = np.array(regSignal) # list to ndarray
```

## ACF and CCF analysis ##

```python
# In this part it is applied a polinomial detrending of the voltage signal, to process a white noise signal to study
# its correlational behaviour

y = y_noise
x = np.arange(0,len(y),1)
sampleNum = len(y)

X1 = np.asarray([np.ones(sampleNum),x]).T
X2 = np.asarray([np.ones(sampleNum),x,x**2]).T
X3 = np.asarray([np.ones(sampleNum),x,x**2,x**3]).T
X4 = np.asarray([np.ones(sampleNum),x,x**2,x**3,x**4]).T
X5 = np.asarray([np.ones(sampleNum),x,x**2,x**3,x**4,x**5]).T

polynoms = [X1,X2,X3,X4,X5]
ratio = 1
w = []
for i in range(5):
    upper = np.dot(y,polynoms[i])
    lower = np.dot(polynoms[i].T,polynoms[i])
    lowerConstant = ratio*np.eye(i+2)
    w.append(np.dot(np.linalg.inv(lower-lowerConstant),upper))
y_trend = np.dot(polynoms[1],w[1])
fsize = 26
fig=plt.figure(101, figsize=(16,4))
plt.plot(y-y_trend)
plt.axis('tight')
plt.xlabel('time [samples]', size = fsize)
plt.ylabel('standardised voltage', size = fsize)
plt.title("Detrended voltage", size=fsize)
plt.show()
no_trend = y-y_trend
```

### ACF y CCF from detrended signal ###

```python
# Plots corresponding to the autocorrelation function of the voltage signal and the cross correlation function between
# voltage and altitude signals. The curves show the third part delay of each correlation.
fsize = 16
n1 = 1000
batch = 1000
signal_autocorr1 = autocorr_test(no_trend[:n1])
signal_xcorr1 = xcorr(no_trend[:n1],a_noise[:n1],batch)
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(2, 2, 1)
plt.plot(signal_autocorr1)
signal_autocorr2 = autocorr_test(no_trend[n1:n1+batch])
signal_xcorr2 = xcorr(no_trend[n1:n1+batch],a_noise[n1:n1+batch],batch)
plt.plot(signal_autocorr2)
signal_autocorr3= autocorr_test(no_trend[n1+batch:n1+2*batch])
signal_xcorr3 = xcorr(no_trend[n1+batch:n1+2*batch],a_noise[n1+batch:n1+2*batch],batch)
plt.plot(signal_autocorr3)
plt.xlabel('lag [sample]',size=fsize)
plt.title('ACF',size=fsize)
plt.rc('xtick', labelsize=fsize)
plt.rc('ytick', labelsize=fsize)
plt.axis('tight')
plt.axis([0,50,-0.6,1])
ax = fig.add_subplot(2, 2, 2)
plt.plot(signal_xcorr1)
plt.plot(signal_xcorr2)
plt.plot(signal_xcorr3)
plt.xlabel('lag [sample]',size=fsize)
plt.title('CCF',size=fsize)
plt.rc('xtick', labelsize=fsize)
plt.rc('ytick', labelsize=fsize)
plt.axis('tight')
plt.axis([0,50,-0.6,0.6])
plt.savefig("acf_ccf.eps",format='eps',  dpi=300, bbox_inches='tight')
plt.show()
```

# Lengthscale optimisation #

##Exhaustive KLMS lengthscale search ##

```python
# Error parameters for training set. In this section, the algorithm is set to use the first 500 samples (data_burn)
# and use all the training samples RMSE prediction to obtain the lengthscales

desiredPercent = 1
percentError = 1-desiredPercent
disp_aux = int(simLength*percentError)
data_burn = 500
```

```python
start_time = time.time()
```

```python
# Exhaustive search array #
perc_array = np.arange(0.1,20,0.05)
```

```python
# KLMS filter parameters #
pred_step = 1
kernel_true = True
gram_matrix = []
kernel_function = gaussian_kernel_multi_channel
learning_rate = np.array([1e-1,0,0,0e-0])
sparsification = 'Novelty'
sparsification_params = np.array([9.5e-1,1e-4,0e-12])
adaptive_kernel = False
adaptive_uncertainty = False
adaptive_params = []
ard_true = False#True
presence_step = 120
hor = 1 # prediction horizon
x_reg = regSignal[1,:-hor,:] # one-channel input
y_del = noiseSignal[1,hor:data_burn] # one-channel output
y = y_del.T
x = x_reg
error6 = []
y = y_del.T
x = x_reg
center_comp = []
# Loop of the exhaustive search #
for i in np.nditer(perc_array):
    perc = i
    sigma = np.array([perc])
    kernel_params = 1/(sigma**2)
    gram_matrix = []#gauss/[]
    # KLMS filter #
    [y_estimate,alpha_history,error,sigma_noise_history,gamma_history,center_num,center_count,center_history,center_index,gram_matrix,runtime,count_arr] = adaptive_filter(x,y,pred_step,\
                                                                                     kernel_true,\
                                                                                     gram_matrix,\
                                                                                     kernel_function,kernel_params,\
                                                                                     delay,\
                                                                                     learning_rate,\
                                                                                     sparsification,\
                                                                                     sparsification_params,\
                                                                                     adaptive_kernel,\
                                                                                     adaptive_uncertainty,\
                                                                                     adaptive_params,ard_true,presence_step)
    div = np.sum(y_noise[disp_aux+hor+delay:]**2)
    num1 = np.sum(error[disp_aux:].T**2)
    error6.append(num1/div)
    center_comp.append(center_num)
```

### Plot of the prediction error for exhaustive search for KLMS ###

```python
# The black curve shows the prediction error for KLMS. It is shown for the array(0,20) with 0.05 of resolution. The red
# cross shows the optimal lengthscale 

perc_inter = np.array(perc_array)
err_inter = np.array(error6)
center_inter = np.array(center_comp)
min_ind = np.where(err_inter[:] == min(err_inter[:]))[0] # minimum of exhaustive search
fsize = 16
fig=plt.figure(101, figsize=(8,4))
ax = plt.gca()
plt.semilogy(perc_inter,err_inter*100,'black',label='KLMS',linewidth=4)
plt.scatter(perc_inter[min_ind], (err_inter[min_ind])*100, s=500, color='red',marker='x',linewidth='5')
plt.xlabel('kernel lengthscale (voltage)', size = fsize)
plt.ylabel('RMSE', size = fsize)
plt.title('Relative MSE', size = fsize)
plt.axis('tight')
plt.tight_layout()
plt.rc('xtick', labelsize=fsize)
plt.rc('ytick', labelsize=fsize)
plt.savefig("kw_sensitivity_1d.eps",format='eps',  dpi=300, bbox_inches='tight')
print("Lengthscale of minimun error: %s"%perc_inter[min_ind])
print("Corresponding support vector: %s"%center_inter[min_ind])
```

### Plot of the dictionary complexity for exhaustive search for KLMS ###

```python
# Plot showing the dictionary complexity for the KLMS search. The black dot shows the dictionary complexity for the 
# optimal point of KLMS

fig=plt.figure(101, figsize=(8,4))
ax = plt.gca()
plt.semilogy(perc_inter,center_inter,'b',label='KLMS',linewidth=4)
plt.semilogy(perc_inter[min_ind],center_inter[min_ind],color='black',marker='o',markersize=15)
plt.xlabel('kernel lengthscale (voltage)', size = fsize)
plt.ylabel('dictionary size', size = fsize)
plt.title('Dictionary complexity', size = fsize)
plt.axis('tight')
plt.rc('xtick', labelsize=fsize)
plt.rc('ytick', labelsize=fsize)
plt.savefig("kw_complexity_1d.eps",format='eps',  dpi=300, bbox_inches='tight')
```

##KLMS-X stochastic lengthscale search ##

```python
# Choose whether the stochastic search is performed or not, the parameters obtained are used in KLMS-X
stochastic_search_bool = False
```

```python
start_time = time.time()
```

```python
if stochastic_search_bool:
    
    sigma_search_vec = []
    sim_num = 10 # number of stochastic runs for the lengthscale 
    
    for k in range(sim_num):
        
        search_length = 200 # number of stochastic steps or iterations of each search
        random_numbers = 1*np.random.sample(search_length)
        random_index = 6*np.random.sample(2)
        lv_0 = random_index[0]
        lh_0 = random_index[-1]
        
        # KLMS-X parameters #
        pred_step = 1
        kernel_true = True
        gram_matrix = []
        kernel_function = gaussian_kernel_multi_channel
        learning_rate = np.array([1e-1,0,0,0e-0])
        sparsification = 'Novelty'
        sparsification_params = np.array([9e-1,1e-4,0e-12]) # Novelty criterion parameters
        adaptive_kernel = False
        adaptive_uncertainty = False
        adaptive_params = []
        ard_true = False#True
        presence_step = 120
        x_reg = regSignal[1:,:-hor,:] # multi-channel input
        y_del = noiseSignal[1:,hor:data_burn] # multi-channel output
        error7 = []
        y = y_del.T
        x = x_reg
        center_comp2 = []
        err_aux = np.inf
        sigma_search = []
        
        for i in range(search_length):
            for j in range(4):
                if (j == 0) and (lv_0 + random_numbers[i] > 0):
                    perc1 = lv_0 + random_numbers[i]
                    perc2 = lh_0
                elif j == 1 and (lh_0 + random_numbers[i] > 0):
                    perc1 = lv_0 
                    perc2 = lh_0 + random_numbers[i]
                elif j == 2 and (lv_0 - random_numbers[i] > 0):
                    perc1 = lv_0 - random_numbers[i]
                    perc2 = lh_0
                elif j == 3 and (lh_0 - random_numbers[i] > 0):
                    perc1 = lv_0 
                    perc2 = lh_0 - random_numbers[i]
                sigma = np.array([perc1,perc2])
                kernel_params = 1/(sigma**2)
                gram_matrix = []
                
                # KLMS-X filter #
                [y_estimate,alpha_history,error,sigma_noise_history,gamma_history,center_num,center_count,center_history,center_index,gram_matrix,runtime,count_arr] = adaptive_filter(x,y,pred_step,\
                                                                                                 kernel_true,\
                                                                                                 gram_matrix,\
                                                                                                 kernel_function,kernel_params,\
                                                                                                 delay,\
                                                                                                 learning_rate,\
                                                                                                 sparsification,\
                                                                                                 sparsification_params,\
                                                                                                 adaptive_kernel,\
                                                                                                 adaptive_uncertainty,\
                                                                                                 adaptive_params,ard_true,presence_step)

                div = np.sum(y_noise[disp_aux+hor+delay:]**2)
                num1 = np.sum(error[disp_aux:,0].T**2)
                if (err_aux > num1/div) and (err_aux >= 0.002):
                    err_aux = num1/div
                    lv_0 = perc1
                    lh_0 = perc2
                    j = 5
                    error7.append(err_aux)
                    sigma_search.append(sigma)
                    center_comp2.append(center_num)
        sigma_search_vec.append(sigma_search)
```

### Plot of the stochastic lengthscale search for KLMS-X ###

```python
#  The black dots show initial conditions and the red crosses show the different run results
if stochastic_search_bool:
    sigma_mean = []
    for i in range(sim_num):
            aux = sigma_search_vec[i]
            aux = np.asarray(aux)
            x = aux[:,0]
            y = aux[:,1]
            fsize = 15
            fig=plt.figure(97, figsize=(8,4))
            plt.scatter(aux[-1,0], aux[-1,1], s=200, color='red',marker='x',linewidth='5')
            plt.scatter(aux[0,0], aux[0,1], s=100, color='black',marker='o',linewidth='5')
            plt.plot(x,y)
            plt.ylabel('kernel lengthscale (altitude)', size = fsize)
            plt.xlabel('kernel lengthscale (voltage)', size = fsize)
            plt.tick_params(axis='both', which='major', labelsize=fsize)
            plt.title("Stochastic minimisation of RMSE (KLMS-X)",size=fsize)#,weight='bold')
            plt.axis('tight')
            sigma_mean.append([aux[-1,0], aux[-1,1]])
    plt.savefig("bayesian_search.eps",format='eps',  dpi=50, bbox_inches='tight')
    plt.show()
    print("Voltage Lengthscale/Altitude Lengthscale: %s"%np.mean(sigma_mean,0))
    print("Voltage Lengthscale Variance/Altitude Lengthscale Variance: %s"%np.var(sigma_mean,0))
```

```python
# Delay array #
del_start = 1
del_step = 1
del_end = 51
delay_array = np.arange(del_start,del_end,del_step)
```

## LMS and KLMS performance for different delays ##

```python
# In this section, the error, runtime and complexity will be computed for the LMS and KLMS algorithm

# Filter parameters #
pred_step = 1
kernel_true = True
gram_matrix = []
kernel_function = gaussian_kernel
sigma0 = 4.4e0 # parameter obtained
learning_rate = np.array([1e-1,0,0,0e-0])
sparsification = 'Novelty'
sparsification_params = np.array([9.5e-1,1e-4,0e-12])
adaptive_kernel = False
adaptive_uncertainty = False
adaptive_params = []
ard_true = False#True
presence_step = 120

error1 = []
err_lms_ar = []
r_time1 = []
r_time3 = []
c_num1 = []
hor = 1
for i in np.nditer(delay_array):
    kernel_true = True
    #print(i)
    delay = i
    #print(delay)
    noiseSignal = [x_noise,y_noise,a_noise]
    noiseSignal = np.array(noiseSignal)
    [noiseSignal,regSignal] = delaySignal(noiseSignal,delay)

    noiseSignal = np.array(noiseSignal)
    regSignal = np.array(regSignal)
    
    # Sigma adjust #
    sigma = sigma0*np.sqrt(delay/30)
    kernel_params = 1/(sigma**2)
    hor = 1
    x_reg = regSignal[1,:-hor,:]
    y_del = noiseSignal[1,hor:]
    x_reg = regSignal[1,:-hor,:]    
    y_del = noiseSignal[1,hor:]
    y = y_del.T
    x = x_reg
    gram_matrix = []
    s_time = time.time()
    learning_rate = np.array([1e-1,0,0,0e-0])
    
    ## KLMS filter ##
    [y_estimate,alpha_history,error,sigma_noise_history,gamma_history,center_num,center_count,center_history,center_index,gram_matrix,runtime,count_arr] = adaptive_filter(x,y,pred_step,\
                                                                                     kernel_true,\
                                                                                     gram_matrix,\
                                                                                     kernel_function,kernel_params,\
                                                                                     delay,\
                                                                                     learning_rate,\
                                                                                     sparsification,\
                                                                                     sparsification_params,\
                                                                                     adaptive_kernel,\
                                                                                     adaptive_uncertainty,\
                                                                                     adaptive_params,ard_true,presence_step)
    r_time1.append(time.time()-s_time)
    s_time = time.time()
    learning_rate = np.array([5e-2,0,0,0e-0])
    kernel_true = False
    
    ## LMS filter ##
    [y_estimate,alpha_history,error_lms_ar,sigma_noise_history,runtime_r] = adaptive_filter(x,y,pred_step,\
                                                                                     kernel_true,\
                                                                                     gram_matrix,\
                                                                                     kernel_function,kernel_params,\
                                                                                     delay,\
                                                                                     learning_rate,\
                                                                                     sparsification,\
                                                                                     sparsification_params,\
                                                                                     adaptive_kernel,\
                                                                                     adaptive_uncertainty,\
                                                                                     adaptive_params,ard_true,presence_step)
    r_time3.append(time.time()-s_time)
    c_num1.append(center_num)
    div = np.sum(y_noise[disp_aux+hor+delay:]**2)
    num1 = np.sum(error[disp_aux:].T**2)
    num3 = np.sum(error_lms_ar[disp_aux:].T**2)
    error1.append(np.sqrt(num1/div))
    err_lms_ar.append(np.sqrt(num3/div))
```

## LMS-X and KLMS-X performance for different delays ##

```python
start_time = time.time()
```

```python
# In this section, the error, runtime and complexity will be computed for the LMS-X and KLMS-X algorithm

# Filter parameters #
pred_step = 1
kernel_true = True
gram_matrix = []#gauss/[]
kernel_function = gaussian_kernel_multi_channel
sigma0 = np.array([6.42706855 , 25.18347832]) # parameters obtained
learning_rate = np.array([1e-1,0,0,0e-0])
sparsification = 'Novelty'
sparsification_params = np.array([9.5e-1,1e-4,0e-12])
adaptive_kernel = False
adaptive_uncertainty = False
adaptive_params = []
ard_true = False
presence_step = 120

error3 = []
err_lms_alt = []
r_time2 = []
r_time4 = []
c_num2 = []
hor = 1
for i in np.nditer(delay_array):
    delay = i
    #print(i)
    noiseSignal = [x_noise,y_noise,a_noise]
    noiseSignal = np.array(noiseSignal) 
    
    [noiseSignal,regSignal] = delaySignal(noiseSignal,delay)

    noiseSignal = np.array(noiseSignal)
    regSignal = np.array(regSignal)
    
    # Sigma adjust #
    sigma = sigma0*np.sqrt(delay/30)
    kernel_params = 1/(sigma**2)
    kernel_true = True
    x_reg = regSignal[1:,:-hor,:]    
    y_del = noiseSignal[1:,hor:]
    y = y_del.T
    x = x_reg
    gram_matrix = []
    s_time = time.time()
    learning_rate = np.array([1e-1,0,0,0e-0])
    ## KLMS-X filter ##
    [y_estimate,alpha_history,error,sigma_noise_history,gamma_history,center_num,center_count,center_history,center_index,gram_matrix,runtime,count_arr] = adaptive_filter(x,y,pred_step,\
                                                                                     kernel_true,\
                                                                                     gram_matrix,\
                                                                                     kernel_function,kernel_params,\
                                                                                     delay,\
                                                                                     learning_rate,\
                                                                                     sparsification,\
                                                                                     sparsification_params,\
                                                                                     adaptive_kernel,\
                                                                                     adaptive_uncertainty,\
                                                                                     adaptive_params,ard_true,presence_step)
    r_time2.append(time.time()-s_time)
    s_time = time.time()
    ## LMS-X filter ##
    kernel_true = False
    learning_rate = np.array([5e-2,0,0,0e-0])
    [y_estimate,alpha_history,error_lms_alt,sigma_noise_history,runtime_r] = adaptive_filter(x,y,pred_step,\
                                                                                     kernel_true,\
                                                                                     gram_matrix,\
                                                                                     kernel_function,kernel_params,\
                                                                                     delay,\
                                                                                     learning_rate,\
                                                                                     sparsification,\
                                                                                     sparsification_params,\
                                                                                     adaptive_kernel,\
                                                                                     adaptive_uncertainty,\
                                                                                     adaptive_params,ard_true,presence_step)
    r_time4.append(time.time()-s_time)
    c_num2.append(center_num)
    div = np.sum(y_noise[disp_aux+hor+delay:]**2)
    num1 = np.sum(error[disp_aux:,0].T**2)
    num3 = np.sum(error_lms_alt[disp_aux:,0].T**2)
    error3.append(np.sqrt(num1/div))
    err_lms_alt.append(np.sqrt(num3/div))
```

```python
# Porcentual adjustment of prediction error and transformation to ndarray #

e_kr_arr = np.array(error1)*100 # Error KLMS
e_ka_arr = np.array(error3)*100 # Error KLMS-X
e_lr_arr = np.array(err_lms_ar)*100 # Error LMS
e_la_arr = np.array(err_lms_alt)*100 # Error LMS-x

t_kr_arr = np.array(r_time1) # Runtime KLMS
t_ka_arr = np.array(r_time2) # Runtime KLMS-X
t_lr_arr = np.array(r_time3) # Runtime LMS
t_la_arr = np.array(r_time4) # Runtime LMS-X

s_kr_arr = np.array(c_num1) # Support Vector KLMS
s_ka_arr = np.array(c_num2) # Support Vector KLMS-X

# Data save in .mat file #
scio.savemat(cwd+' results', mdict={'e_kr_arr': e_kr_arr,\
                                       'e_ka_arr': e_ka_arr,\
                                       'e_lr_arr': e_lr_arr,\
                                       'e_la_arr': e_la_arr,\
                                       't_kr_arr': t_kr_arr,\
                                       't_ka_arr': t_ka_arr,\
                                       't_lr_arr': t_lr_arr,\
                                       't_la_arr': t_la_arr,\
                                       's_kr_arr': s_kr_arr,\
                                       's_ka_arr': s_ka_arr})
```
